Data 1

```python
import matplotlib.pyplot as plt
import numpy as np

time_amp = np.genfromtxt('Cassy_data_1.txt',dtype='str')

t = np.array(time_amp[:,1])
time = np.asarray(t, dtype=float)

a = np.array(time_amp[:,2])
amp = np.asarray(a, dtype=float)

ampl = amp - np.mean(amp)

plt.plot(time, ampl)
plt.xlabel("Time (s)", fontsize = 12)
plt.ylabel("Amplitude", fontsize = 12)
plt.title("Graph of Amplitude vs Time \n No Damping No Weight ", fontsize = 16)
plt.grid()
```
